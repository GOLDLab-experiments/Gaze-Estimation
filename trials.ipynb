```python
from ultralytics import YOLO

model_name = "yolo11s50"

# Load a pretrained YOLO model
model = YOLO(f"trained_models/{model_name}.pt")

results = model.predict("helmet/images/train/helm_000010.jpg", conf = 0.25, save = False, save_dir = f"{model_name}_output")
```

```python
detections = results[0].boxes.data.tolist()
```
